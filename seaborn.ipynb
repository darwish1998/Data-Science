```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('dm_office_sales.csv')
data.head()
```

```python
plt.figure(figsize=(12,4))
sns.scatterplot(x='salary',y='sales', data=data, hue='level of education')
```

```python
sns.scatterplot(x='salary',y='sales', data=data, hue='level of education',style='level of education' )
# platete documentation
```

```python
sns.displot(data=data, x='salary',bins=20, kde=True, rug=True)
sns.set(style='dark')
```

```python
data['division'].value_counts()
```

```python
plt.figure(figsize=(10,3),dpi=100)
sns.countplot(data=data,x='level of education', hue='division', palette='Set2')
```

```python
data = pd.read_csv('StudentsPerformance.csv')
data.head()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(data=data,y='math score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1, .7))
```

```python
plt.figure(figsize=(15,9))
sns.violinplot(data=data,y='reading score', x='parental level of education', hue='test preparation course', split='True')
plt.legend(bbox_to_anchor=(1, .7))
```

```python
sns.swarmplot(data=data, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.0,1))
```

```python
plt.figure(figsize=(10,10))
sns.boxenplot(data=data,y='math score', x='parental level of education', hue='test preparation course')
plt.legend(bbox_to_anchor=(1, .7))
```

```python
sns.jointplot(data=data , x='math score', y = 'reading score', kind='kde', shade='True')
```

```python
sns.jointplot(data=data , x='math score', y = 'reading score', hue='gender')
```

```python
sns.pairplot(data=data,hue='gender', corner=True)
```

```python
sns.catplot(data=data, x='gender', y='math score', kind='violin', col='lunch', row='test preparation course')
```
